# Index Rename

**We can rename index or column names**
* Mapper + Axis
* Index + Column

**Mapper + Axis**

```python
import pandas as pd
```

```python
a = pd.read_csv('/kaggle/input/data-science-jobs-in-india/naukri_data_science_jobs_india.csv',index_col = ['Job_Role'])
a
```

```python
a
```

```python
a.rename(mapper = {"Senior Data Scientist":"Senior Data Analyst",
                   "Data Scientist":"Data Analyst"},
                    axis = 0, inplace = True)
```

```python
a.rename(mapper = {"Senior Data Scientist":"Senior Data Analyst",
                   "Data Scientist":"Data Analyst"},
                    axis = "rows", inplace = True) #axis =0/index/rows means the same
a
```

**Using index and column argument**

```python
a
```

```python
a.rename(index = {"Senior Data Analyst":"Senior Data Scientist",
                 "Data Analyst":"Data Scientist"}, inplace = True)
a
```

# Column Renaming

**Use mapper and axis**

```python
import pandas as pd
```

```python
a = pd.read_csv('/kaggle/input/data-science-jobs-in-india/naukri_data_science_jobs_india.csv', index_col = ['Job_Role'])
a
```

```python
a.rename(mapper = {"Company" : "Firm",
                  "Skills/Description" : "Jor Requirement"}, axis = 1)

#To change the name of the column
```

```python
a #no permanent change observed
```

```python
a.rename(mapper = {"Company" : "Firm",
                  "Skills/Description" : "Jor Requirement"}, axis = 1, inplace = True)

#To change the name of the column and keeping it as a permanent change
a
```

```python
a.columns
```

```python
a
```

```python
a.rename(columns = {"Firm": "Company",
                   "Jor Requirement": "Skills/Description"}, inplace = True)
a
```
